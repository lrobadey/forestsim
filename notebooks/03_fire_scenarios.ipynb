# Fire Scenarios

Append a demo fire event and compare the edited branch against the baseline run.


```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wattforest import LandscapeInitializer
from wattforest.events import EventType, SimEvent
from wattforest.metrics import PatternMetrics

MANIFEST = Path('../data/sites/synthetic_demo/site_manifest.json')
engine = LandscapeInitializer.from_site_manifest(MANIFEST)
fire = SimEvent(event_id='demo-fire', event_type=EventType.FIRE_IGNITION, year=2020, params={'ignition_cells': [(40, 70)], 'duration_hr': 8.0, 'wind_speed_ms': 10.0, 'wind_dir_deg': 85.0})
engine.event_log.events.append(fire)
engine.run(2015, 2025)
print('burned area ha', round(max(r.area_burned_ha for r in engine.history), 2))
```
