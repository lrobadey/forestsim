# Scenario Comparison

Compare the baseline demo branch with a fire-edited branch using the same manifest package.


```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wattforest import LandscapeInitializer
from wattforest.events import EventType, SimEvent
from wattforest.metrics import PatternMetrics

MANIFEST = Path('../data/sites/synthetic_demo/site_manifest.json')
baseline = LandscapeInitializer.run_phase3_baseline(MANIFEST, end_year=2025)
branch = LandscapeInitializer.from_site_manifest(MANIFEST)
branch.event_log.events.append(SimEvent(event_id='fire-compare', event_type=EventType.FIRE_IGNITION, year=2021, params={'ignition_cells': [(35, 65)], 'duration_hr': 6.0, 'wind_speed_ms': 12.0, 'wind_dir_deg': 90.0}))
branch.run(2015, 2025)
series = pd.DataFrame({'year': [r.year for r in baseline.engine.history], 'baseline': [r.total_biomass_kg for r in baseline.engine.history], 'branch': [r.total_biomass_kg for r in branch.history]})
print(series.tail())
```
