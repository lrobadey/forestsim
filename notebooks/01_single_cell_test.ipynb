# Single Cell Test

Inspect one demo cell and verify the age-class distribution helper.


```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wattforest import LandscapeInitializer
from wattforest.events import EventType, SimEvent
from wattforest.metrics import PatternMetrics

MANIFEST = Path('../data/sites/synthetic_demo/site_manifest.json')
engine = LandscapeInitializer.from_site_manifest(MANIFEST)
cell = engine.vegetation[50, 60]
print('cohorts', len(cell.cohorts), 'cover', round(cell.total_canopy_cover, 3), 'mean_age', round(cell.mean_age, 1))
grid = np.empty(engine.config.shape, dtype=object)
for r in range(engine.config.shape[0]):
    for c in range(engine.config.shape[1]):
        grid[r, c] = engine.vegetation[r, c]
print(PatternMetrics.age_class_distribution(grid))
```
