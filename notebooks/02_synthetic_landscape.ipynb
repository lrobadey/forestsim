# Synthetic Landscape

Plot the synthetic terrain, climate, and canopy structure used by the demo package.


```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wattforest import LandscapeInitializer
from wattforest.events import EventType, SimEvent
from wattforest.metrics import PatternMetrics

MANIFEST = Path('../data/sites/synthetic_demo/site_manifest.json')
engine = LandscapeInitializer.from_site_manifest(MANIFEST)
fig, axes = plt.subplots(1, 3, figsize=(14, 4), constrained_layout=True)
axes[0].imshow(engine.terrain.elevation, cmap='terrain')
axes[0].set_title('DEM')
axes[1].imshow(engine.climate.growing_degree_days, cmap='YlOrRd')
axes[1].set_title('Baseline GDD')
axes[2].imshow(engine.canopy_cover_grid(), cmap='Greens', vmin=0, vmax=1)
axes[2].set_title('Canopy cover')
for ax in axes: ax.set_axis_off()
plt.show()
```
